```python
import pandas as pd
import numpy as np

# Correct file path using a raw string (r"") or double backslashes
file_path = r"C:\Users\MOURIYA DAMPETLA\Downloads\Reviews (1).csv"

# Read CSV file
reviews_datasets = pd.read_csv("C:\\Users\\MOURIYA DAMPETLA\\Downloads\\Reviews (1).csv")

# Select the first 20,000 rows
reviews_datasets = reviews_datasets.head(20000)

# Drop missing values
reviews_datasets = reviews_datasets.dropna()  # OR use reviews_datasets.dropna(inplace=True)

# Display first few rows
print(reviews_datasets.head())
```

```python
reviews_datasets['Text'][350]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

 

count_vect = CountVectorizer(max_df=0.8, min_df=2, stop_words='english')

doc_term_matrix = count_vect.fit_transform(reviews_datasets['Text'].values.astype('U'))
```

```python
print(doc_term_matrix)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=5, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
import random
for i in range(10):
    random_id = random.randint(0,len(count_vect.get_feature_names_out()))
    print(count_vect.get_feature_names_out()[random_id])
```

```python
first_topic = LDA.components_[0] # topic 0
```

```python
top_topic_words = first_topic.argsort()[-10:]
```

```python
for i in top_topic_words:
    print(count_vect.get_feature_names_out()[i])
```

```python
for i,topic in enumerate(LDA.components_):

    print(f'Top 10 words for topic #{i}:')

    print([count_vect.get_feature_names_out()[i] for i in topic.argsort()[-10:]])

    print('\n')
```
